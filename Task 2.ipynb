# Visualization of 80 Cereals

```python
plt.figure(figsize=(8,5))
plt.title("Cereals of data distribution")
sns.histplot(df) # Here fat cereals has a highest distribution among all others
```

```python
def plot_histogram(column_data,column_name,color):
    plt.figure(figsize=(5,3))
    plt.title(f"Distribution of {column_name}")
    sns.histplot(column_data,kde=True,color='red')
    plt.show()
```

```python
plot_histogram(df['mfr'],'Manufacturer',color='red')
```

```python
plot_histogram(df['type'],'Cold or Hot Cereals',color='red') # HERE COLD TYPE OF CEREALS HAS A HIGH DISTRIBUTION
```

```python
plot_histogram(df['calories'],'Calories',color='red') # Here the high distribution of calories range between 100-120 calories
```

```python
plot_histogram(df['protein'],'protein',color='red') # The distribution of protein falls within the range between falls 2-3 grams on x-axis
```

```python
plot_histogram(df['sodium'],'sodium',color='red') # We observe that our standard distribution ranges between 150-250 increases and slightly decrease in 300
```

```python
plot_histogram(df['carbo'],'carbo',color='red') # HERE THE CARBO RATE OF PROTEINS INCREASES THE CARBO COUNT FROM 15.3 AND SLIGHTLY DECREASES FROM 13 CARBO COUNTS
```

```python
plot_histogram(df['sugars'],'sugars',color='red') # here the count of sugars has 4 which is largest count of 13 among all others
```

# WHICH BRAND HAS HIGH RATINGS

```python
cereals_sorting = df.sort_values(by='rating',ascending=False)
plt.figure(figsize=(15,3))
plt.title("Brands and their ratings")
plt.xticks(rotation=90)
sns.barplot(data = cereals_sorting, x=cereals_sorting['name'], y=cereals_sorting['rating'])
plt.show() # ALL BRAN WITH EXTRA FIBER HAS HIGHEST DISTRIBUTION RATINGS
```

# RELATIONSHIP BETWEEN SUGAR AND RATINGS

```python
plt.figure(figsize=(10,5))
plt.title("Relationship between sugar and ratings")
sns.scatterplot(data=df,x=df['sugars'],y=df['rating'])
plt.show() # IT HAS LOWER SUGAR CONTENT AND HIGHER RATING IN THE CEREALS
```
